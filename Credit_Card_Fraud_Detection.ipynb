### Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset 
CreditCard_data = pd.read_csv("creditcard.csv")

# Show the first 5 rows
CreditCard_data.head()
```

```python
# The shape of the data
CreditCard_data.shape
```

```python
# Inforamtions about the data
CreditCard_data.info()
```

```python
# The number of normal and fraud transactions

### 0 --> normal transaction
### 1 --> fraud transaction

CreditCard_data["Class"].value_counts()
```

### Separating the data for analysis

```python
legit = CreditCard_data[CreditCard_data["Class"] == 0]
fraud = CreditCard_data[CreditCard_data["Class"] == 1]
```

```python
print(f"legit data shape:{legit.shape}\nfraud data shape:{fraud.shape}")
```

##### This is unbalanced data because there are many labels in legit class and few data in fraud class  

```python
# Statistical measures of the legit data
legit.describe()
```

```python
# Statistical measures of the fraud data
fraud.describe()
```

```python
# Getting the mean of features for each Transactions
CreditCard_data.groupby("Class").mean()
```

#### Under-sampling

```python
# Take a sample from legit dataset have the same distribution
legit_sample = legit.sample(n = 492)
```

```python
### Concatenate the two classe
sample_data = pd.concat([legit_sample,fraud] , axis = 0)
sample_data.head()
```

```python
# The number of normal and fraud transactions in the new data
sample_data["Class"].value_counts()
```

```python
sample_data.groupby(['Class']).mean()
```

```python
# shuffling the dataset
sample_data = sample_data.sample(frac = 1)
```

### Splitting the dataset


```python
X = sample_data.drop("Class" , axis = 1)
Y = sample_data["Class"]
X.shape , Y.shape
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.2 , random_state=1)

x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

### Make  a Logistic Regression Model

```python
model = LogisticRegression()
```

```python
# Training the model
model.fit(x_train , y_train)
```

### Model Evaluation

```python
# Accuracy on training data
x_train_pred = model.predict(x_train)
training_acc = accuracy_score(x_train_pred , y_train)
print(f"The accuracy on training data: {np.round(training_acc , 2)*100}%")
```

```python
# Accuracy on training data
x_test_pred = model.predict(x_test)
testing_acc = accuracy_score(x_test_pred , y_test)
print(f"The accuracy on testing data: {np.round(testing_acc , 2)*100}%")
```
